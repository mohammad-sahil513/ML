````python
from ucimlrepo import fetch_ucirepo

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets

# metadata
print(iris.metadata)

# variable information
print(iris.variables)
````

The Naive Bayes classification and evaluation have been completed in the cells above.

### Split the dataset into training and testing sets

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Flatten y_train and y_test for use with scikit-learn models
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()
````

### Train and Evaluate Gaussian Naive Bayes

````python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

gaussian_nb = GaussianNB()
gaussian_nb.fit(X_train, y_train)
gaussian_predictions = gaussian_nb.predict(X_test)

accuracy = accuracy_score(y_test, gaussian_predictions)
precision = precision_score(y_test, gaussian_predictions, average='weighted')
recall = recall_score(y_test, gaussian_predictions, average='weighted')
f1 = f1_score(y_test, gaussian_predictions, average='weighted')
conf_matrix = confusion_matrix(y_test, gaussian_predictions)

print("Gaussian Naive Bayes Results:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F-score: {f1:.4f}")
print(f"Confusion Matrix:\n{conf_matrix}")
````

### Train and Evaluate Multinomial Naive Bayes

````python
from sklearn.naive_bayes import MultinomialNB

multinomial_nb = MultinomialNB()
multinomial_nb.fit(X_train, y_train)
multinomial_predictions = multinomial_nb.predict(X_test)

multinomial_accuracy = accuracy_score(y_test, multinomial_predictions)
multinomial_precision = precision_score(y_test, multinomial_predictions, average='weighted')
multinomial_recall = recall_score(y_test, multinomial_predictions, average='weighted')
multinomial_f1 = f1_score(y_test, multinomial_predictions, average='weighted')
multinomial_conf_matrix = confusion_matrix(y_test, multinomial_predictions)

print("\nMultinomial Naive Bayes Results:")
print(f"Accuracy: {multinomial_accuracy:.4f}")
print(f"Precision: {multinomial_precision:.4f}")
print(f"Recall: {multinomial_recall:.4f}")
print(f"F-score: {multinomial_f1:.4f}")
print(f"Confusion Matrix:\n{multinomial_conf_matrix}")
````

### Train and Evaluate Multinomial Naive Bayes(Parameter Tuning)

````python
from sklearn.naive_bayes import MultinomialNB

multinomial_nb = MultinomialNB(alpha=2.5, fit_prior=True, class_prior=None).fit(X_train, y_train)
multinomial_nb.fit(X_train, y_train)
multinomial_predictions = multinomial_nb.predict(X_test)

multinomial_accuracy = accuracy_score(y_test, multinomial_predictions)
multinomial_precision = precision_score(y_test, multinomial_predictions, average='weighted')
multinomial_recall = recall_score(y_test, multinomial_predictions, average='weighted')
multinomial_f1 = f1_score(y_test, multinomial_predictions, average='weighted')
multinomial_conf_matrix = confusion_matrix(y_test, multinomial_predictions)

print("\nMultinomial Naive Bayes Results:")
print(f"Accuracy: {multinomial_accuracy:.4f}")
print(f"Precision: {multinomial_precision:.4f}")
print(f"Recall: {multinomial_recall:.4f}")
print(f"F-score: {multinomial_f1:.4f}")
print(f"Confusion Matrix:\n{multinomial_conf_matrix}")
````

### Train and Evaluate Bernoulli Naive Bayes

````python
from sklearn.naive_bayes import BernoulliNB

bernoulli_nb = BernoulliNB()
bernoulli_nb.fit(X_train, y_train)
bernoulli_predictions = bernoulli_nb.predict(X_test)

bernoulli_accuracy = accuracy_score(y_test, bernoulli_predictions)
bernoulli_precision = precision_score(y_test, bernoulli_predictions, average='weighted', zero_division=0)
bernoulli_recall = recall_score(y_test, bernoulli_predictions, average='weighted')
bernoulli_f1 = f1_score(y_test, bernoulli_predictions, average='weighted')
bernoulli_conf_matrix = confusion_matrix(y_test, bernoulli_predictions)

print("\nBernoulli Naive Bayes Results:")
print(f"Accuracy: {bernoulli_accuracy:.4f}")
print(f"Precision: {bernoulli_precision:.4f}")
print(f"Recall: {bernoulli_recall:.4f}")
print(f"F-score: {bernoulli_f1:.4f}")
print(f"Confusion Matrix:\n{bernoulli_conf_matrix}")
````

## Train decision tree

### Subtask:
Train a Decision Tree classifier on the training data.


**Reasoning**:
Import the Decision Tree classifier and train the model on the training data.



````python
from sklearn.tree import DecisionTreeClassifier

# Instantiate a DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier(random_state=42)

# Train the model
dt_classifier.fit(X_train, y_train)
````

## Evaluate decision tree

### Subtask:
Evaluate the Decision Tree model on the testing data and display the classification results (Accuracy, Precision, Recall, F-score, confusion matrix).


**Reasoning**:
Evaluate the trained Decision Tree model using the test set and calculate performance metrics.



````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Make predictions
y_pred_dt = dt_classifier.predict(X_test)

# Calculate evaluation metrics
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt, average='weighted')
recall_dt = recall_score(y_test, y_pred_dt, average='weighted')
f1_dt = f1_score(y_test, y_pred_dt, average='weighted')
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)

# Store metrics
decision_tree_metrics_gini = {
    'Accuracy': accuracy_dt,
    'Precision': precision_dt,
    'Recall': recall_dt,
    'F1-score': f1_dt,
    'Confusion Matrix': conf_matrix_dt
}
decision_tree_model = dt_classifier
# Display the metrics
print("Decision Tree (Gini) Metrics:")
for metric, value in decision_tree_metrics_gini.items():
    print(f"{metric}:")
    display(value)
````

## Visualize decision tree (gini)

### Subtask:
Generate and display a visualization of the Decision Tree highlighting information gain using the Gini impurity.


**Reasoning**:
Generate and display a visualization of the Decision Tree highlighting information gain using the Gini impurity.



````python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(20, 10))
plot_tree(decision_tree_model,
          feature_names=X.columns.tolist(),
          class_names=decision_tree_model.classes_,
          filled=True,
          impurity=True,
          rounded=True)
plt.title("Decision Tree Visualization (Gini Impurity)")
plt.show()
````

## Visualize decision tree (entropy)

### Subtask:
Generate and display a visualization of the Decision Tree highlighting information gain using Entropy.


**Reasoning**:
Generate and display a visualization of the Decision Tree highlighting information gain using Entropy.



````python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Create a new Decision Tree instance with entropy criterion
decision_tree_entropy = DecisionTreeClassifier(criterion='entropy', random_state=42)

# Train the model
decision_tree_entropy.fit(X_train, y_train)

# Set the figure size for the plot
plt.figure(figsize=(20, 10))

# Visualize the decision tree with entropy
plot_tree(decision_tree_entropy,
          feature_names=X.columns.tolist(),
          class_names=decision_tree_entropy.classes_,
          filled=True,
          impurity=True,
          rounded=True,
          proportion=True)

# Add a title to the plot
plt.title("Decision Tree Visualization (Entropy)")

# Display the plot
plt.show()
````

*   Visualizations of the Decision Tree were successfully generated highlighting both Gini impurity and Entropy as the splitting criteria.

### Insights or Next Steps
